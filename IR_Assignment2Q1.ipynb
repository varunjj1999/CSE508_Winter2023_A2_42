```python
import math
def cosine_similarity(v1, v2):
    dot_product = 0
    norm_v1 = 0
    norm_v2 = 0
    for i in range(len(v1)):
        dot_product += v1[i] * v2[i]
        norm_v1 += v1[i] ** 2
        norm_v2 += v2[i] ** 2
    norm_v1 = math.sqrt(norm_v1)
    norm_v2 = math.sqrt(norm_v2)
    similarity = dot_product / (norm_v1 * norm_v2)
    return similarity
```

```python
def top_5_indexes(lst):
  indexed_lst = list(enumerate(lst))
  sorted_lst = sorted(indexed_lst, key=lambda x: x[1], reverse=True)
  top_5_indexes = [x[0] for x in sorted_lst[:5]]
  return top_5_indexes
```
